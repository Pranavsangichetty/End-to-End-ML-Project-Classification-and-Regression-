```python
clf_target = 'Converted'  

if 'lead_score' in df.columns:
    reg_target = 'lead_score'
else:
    reg_target = 'synthetic_lead_score'
    score = np.zeros(len(df))
    if 'TotalVisits' in df.columns:
        score += (df['TotalVisits'].fillna(0).rank(pct=True).values)
    if 'Page Views Per Visit' in df.columns:
        score += (df['Page Views Per Visit'].fillna(0).rank(pct=True).values)
    score = 100 * (score - score.min())/(score.max() - score.min() + 1e-9)
    df[reg_target] = score
```

**Feature Selection**

```python
X = df.drop(columns=[clf_target, reg_target]) if reg_target in df.columns else df.drop(columns=[clf_target])
drop_cols = [c for c in X.columns if 'id' in c.lower() or 'number' in c.lower() or X[c].nunique()==len(X)]
X = X.drop(columns=[c for c in drop_cols if c in X.columns])
print("Dropped id-like:", drop_cols)
```

## Classification Models

**Step-1: Split the Data**

```python
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()
print(len(num_cols), "numeric cols;", len(cat_cols), "categorical cols")
```

**Step-2: Preprocessing Pipeline**

```python
num_cols = X.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols = X.select_dtypes(exclude=['int64','float64']).columns.tolist()

low_card = [c for c in cat_cols if X[c].nunique() <= 20]
high_card = [c for c in cat_cols if X[c].nunique() > 20]

print("Low-cardinality categorical:", low_card)
print("High-cardinality categorical:", high_card)
```

```python
numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_low_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='MISSING')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

cat_high_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='MISSING')),
    ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_cols),
        ('low', cat_low_transformer, low_card),
        ('high', cat_high_transformer, high_card)
    ],
    remainder='drop'
)
```

**Step-3: Train Test Split**

```python
X_clf = X.copy()
y_clf = df[clf_target].astype(int)
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.25, random_state=42, stratify=y_clf)
```

**Step-4: Classification Models**

```python
models_clf = {
    "LogisticRegression": Pipeline([('pre', preprocessor), ('clf', LogisticRegression(max_iter=2000))]),
    "RandomForest": Pipeline([('pre', preprocessor), ('clf', RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1))]),
    "AdaBoost": Pipeline([('pre', preprocessor), ('clf', AdaBoostClassifier(n_estimators=100, random_state=42))])
}
```

**Step-5: Classification Model Evaluation**

```python
results_clf = {}
for name, pipe in models_clf.items():
    pipe.fit(X_train_clf, y_train_clf)
    y_pred = pipe.predict(X_test_clf)
    y_proba = pipe.predict_proba(X_test_clf)[:,1] if hasattr(pipe.named_steps['clf'], "predict_proba") else None
    metrics = {
        "accuracy": accuracy_score(y_test_clf, y_pred),
        "precision": precision_score(y_test_clf, y_pred, zero_division=0),
        "recall": recall_score(y_test_clf, y_pred, zero_division=0),
        "f1": f1_score(y_test_clf, y_pred, zero_division=0),
        "roc_auc": roc_auc_score(y_test_clf, y_proba) if y_proba is not None else np.nan
    }
    results_clf[name] = metrics
    print(name, metrics)
```

## Regression Models

**Step-1: Train and Test Split**

```python
y_reg = df[reg_target]
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.25, random_state=42)
```

**Step-2: Regression Models**

```python
from sklearn.neighbors import KNeighborsRegressor
models_reg = {
    "LinearRegression": Pipeline([('pre', preprocessor), ('reg', LinearRegression())]),
    "Ridge": Pipeline([('pre', preprocessor), ('reg', Ridge())]),
    "RandomForestRegressor": Pipeline([('pre', preprocessor), ('reg', RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1))]),
    "KNN": Pipeline([('pre', preprocessor), ('reg', KNeighborsRegressor(n_neighbors=5))])
}
```

**Step-3: Regression Model Evaluation**

```python
results_reg = {}
for name, pipe in models_reg.items():
    pipe.fit(X_train_reg, y_train_reg)
    y_pred = pipe.predict(X_test_reg)
    metrics = {
        "r2": r2_score(y_test_reg, y_pred)
    }
    results_reg[name] = metrics
    print(name, metrics)
```
